```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# 2. إعادة تسمية الأعمدة لإزالة الرموز غير القياسية
df.columns = df.columns.str.replace(r'\(?C\)?', 'C', regex=True)  # استبدال ?C بـ C
df.columns = df.columns.str.replace(r'\(?m\)?', 'm', regex=True)  # استبدال ?m بـ m
df.columns = df.columns.str.replace(r'\?', '', regex=True)  # إزالة أي علامات استفهام زائدة
df.columns = df.columns.str.replace('_Te(m)perature', '_Temperature', regex=True)  # تصحيح Temperature
df.columns = df.columns.str.replace('_Syste(m)', '_System', regex=True)  # تصحيح System
df.columns = df.columns.str.replace(r'Torque\(Nm', 'Torque(Nm)', regex=True)  # تصحيح Torque
df.columns = df.columns.str.replace(r'Hydraulic_Oil_TemperatureC', 'Hydraulic_Oil_Temperature(C)', regex=True)
df.columns = df.columns.str.replace(r'Spindle_Bearing_TemperatureC', 'Spindle_Bearing_Temperature(C)', regex=True)
df.columns = df.columns.str.replace(r'Spindle_Vibrationm', 'Spindle_Vibration(m)', regex=True)
df.columns = df.columns.str.replace(r'Tool_Vibrationm', 'Tool_Vibration(m)', regex=True)

df.rename(columns={'Downti(m)e': 'Downtime'}, inplace=True)  # تصحيح Downtime

# 3. التحقق من القيم الفارغة ومعالجتها
print("Missing values before handling:\n", df.isnull().sum())

# ملء القيم الفارغة في Machine_ID و Assembly_Line_No ب Unknown
df['Machine_ID'] = df['Machine_ID'].fillna('Unknown')
df['Assembly_Line_No'] = df['Assembly_Line_No'].fillna('Unknown')

# ملء القيم الفارغة في Downtime بالقيمة 0 (إذا كان الغياب يعني عدم حدوث توقف)
df['Downtime'] = df['Downtime'].fillna(0)

# التأكد من أن الأعمدة التصنيفية تبقى كنصوص
df['Machine_ID'] = df['Machine_ID'].astype(str)
df['Assembly_Line_No'] = df['Assembly_Line_No'].astype(str)

# تحديد الأعمدة الرقمية فقط لتطبيق median()
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())  # تعويض القيم الناقصة بالوسيط

# 4. التأكد من أن الأعمدة الرقمية مسجلة بشكل صحيح
non_numeric_cols = ['Machine_ID', 'Assembly_Line_No']  # الاحتفاظ بهذه الأعمدة كـ Categorical
df[list(set(df.columns) - set(non_numeric_cols))] = df[list(set(df.columns) - set(non_numeric_cols))].apply(pd.to_numeric, errors='coerce')

# 5. تحليل التكرارات بدلاً من حذفها مباشرة
duplicate_rows = df[df.duplicated()]
print("Number of duplicate rows:", duplicate_rows.shape[0])
if not duplicate_rows.empty:
    print("Duplicate rows found! Consider reviewing them before deletion.")

# 6. حفظ البيانات النظيفة
df.to_csv("Cleaned_Machine_Downtime.csv", index=False)

# طباعة الأعمدة بعد التصحيح
print("Updated columns:\n", df.columns)

print("Data cleaning completed and saved as 'Cleaned_Machine_Downtime.csv'")
```

```python
try:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    if df['Date'].dtype in ['int64', 'float64']:
        df['Date'] = pd.to_datetime(df['Date'], unit='ns', errors='coerce')
except Exception as e:
    print(f"Error in date conversion: {e}")

# 2. إعادة تسمية الأعمدة لإزالة الرموز غير القياسية
replacement_dict = {
    r'\(?C\)?': 'C',
    r'\(?m\)?': 'm',
    r'\?': '',
    '_Te\(m\)perature': '_Temperature',
    '_Syste\(m\)': '_System',
    r'Torque\(Nm': 'Torque(Nm)',
    r'Hydraulic_Oil_Temperature\(C': 'Hydraulic_Oil_Temperature(C)',
    r'Spindle_Bearing_Temperature\(C': 'Spindle_Bearing_Temperature(C)',
    r'Spindle_Vibration\(m': 'Spindle_Vibration(m)',
    r'Tool_Vibration\(m': 'Tool_Vibration(m)'
}

for pattern, replacement in replacement_dict.items():
    df.columns = df.columns.str.replace(pattern, replacement, regex=True)

df.rename(columns={'Downti(m)e': 'Downtime'}, inplace=True)

# 3. التحقق من القيم الفارغة ومعالجتها
print("Missing values before handling:\n", df.isnull().sum())

# تعويض القيم الفارغة في الأعمدة النصية بالقيم الأكثر شيوعًا
for col in ['Machine_ID', 'Assembly_Line_No']:
    if df[col].isnull().any():
        df[col].fillna(df[col].mode()[0], inplace=True)

if df['Downtime'].isnull().any():
    df['Downtime'].fillna(df['Downtime'].median(), inplace=True)

# تحديد الأعمدة الرقمية فقط لتطبيق median()
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# 4. التأكد من أن الأعمدة الرقمية مسجلة بشكل صحيح
object_cols = df.select_dtypes(include=['object']).columns
df[object_cols] = df[object_cols].apply(pd.to_numeric, errors='coerce')

# 5. إزالة أي صفوف مكررة
df = df.drop_duplicates()

df.to_csv("Cleaned_Machine_Downtime.csv", index=False)

print("Updated columns:\n", df.columns)

print("Data cleaning completed and saved as 'Cleaned_Machine_Downtime.csv'")
```

```python
import matplotlib.pyplot as plt

# ترتيب البيانات حسب التاريخ
df = df.sort_values("Date")

fig, ax1 = plt.subplots(figsize=(12, 6))

# تطبيق Moving Average لتخفيف التذبذب أكثر
df["Smoothed_Oil_Temp"] = df["Hydraulic_Oil_TemperatureC"].rolling(window=40).mean()
df["Smoothed_Spindle_Speed"] = df["Spindle_Speed(RPM)"].rolling(window=50).mean()

# رسم درجة حرارة الزيت باللون الأزرق
ax1.plot(df["Date"], df["Smoothed_Oil_Temp"], 
         color="blue", label="Oil Temp (Smoothed)", linewidth=2.5)
ax1.set_ylabel("Oil Temp (°C)", color="blue", fontsize=14)
ax1.tick_params(axis='y', labelcolor="blue")

# إضافة محور ثاني لسرعة المغزل
ax2 = ax1.twinx()
ax2.plot(df["Date"], df["Smoothed_Spindle_Speed"], 
         color="darkorange", alpha=0.85, label="Spindle Speed (Smoothed)", 
         linewidth=3, linestyle="dashed")
ax2.set_ylabel("Spindle Speed (RPM)", color="darkorange", fontsize=14)
ax2.tick_params(axis='y', labelcolor="darkorange")

# تحسين شكل المحاور والعناوين
plt.title("Hydraulic Oil Temp & Spindle Speed Over Time", fontsize=20, fontweight="bold")
ax1.set_xlabel("Date", fontsize=14)

# تدوير التواريخ لتحسين الوضوح
plt.xticks(rotation=45, ha='right')

# إضافة خطوط الشبكة
ax1.grid(True, linestyle="--", alpha=0.5)
ax2.grid(False)

# تحسين توزيع المسافات في الرسم
fig.tight_layout()

plt.show()
```
